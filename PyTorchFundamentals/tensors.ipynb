### Random Tensors

why random tensors?

The way that Neural Network learn is that they start with tensors full of random values and adjust those random numbers to represent the data better 

`Start with random nums -> look at data -> update random numbers -> repeat till some condition or desired output`

```python
# Creating Random Tensors 

random_tensor = torch.rand(3, 4)
random_tensor
```

```python
# Creating random image tensors with the size of image tensor
random_image_size_tensor = torch.rand(size=(224, 224, 3)) # height, width, colour channel 

random_image_size_tensor.shape, random_image_size_tensor.ndim
```

```python
# Matrix Multiplication - Dot product

tensor_1 = torch.tensor([
    [1, 2, 3],
    [5, 3, 6],
    [12, 3, 4]
])
tensor_1 * tensor_1
```

```python
# %%time

# Matrix Multiplication - Cross Product
torch.matmul(tensor_1, tensor_1)
```

```python
# %%time

len(tensor_1)

res_tensor = torch.zeros(len(tensor_1), len(tensor_1))

for i in range(len(tensor_1)):
    for j in range(len(tensor_1[0])):
        for k in range(len(tensor_1)):
            res_tensor[i][j] += tensor_1[i][k] * tensor_1[k][j]

print(res_tensor)
```

### Reshaping, Stacking, squeezing, unsqueezing tensors 

* Reshaping Tensor - reshapes an input tensor to a defined shape
* View - Returns of a view of an input tensor which shares the same memory
* Stacking - Concatenates the tensors on top of each other (**Vstack**) or side by side (**Hstack**)
* Squeeze - remove all `1` dimentions from a tensor
* Unsqueeze - add a `1` dimention to a target tensor
* Commute - return a view of a input with dimentions permuted(swapped) in a certain way

```python
x = torch.arange(1., 11.)

x, x.shape
```

```python
# reshape

x.reshape((2, 5))
```

```python
# change the view 

z = x.view(1, 10)
z, z.shape
```

```python
# Stack tensors on top of each other 

x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# Squeeze - removes the all the single dimentions from (1, 223, 323) -> (223, 323)

x_squeezed = torch.squeeze(x)
x_squeezed
```

```python
# unsqueeze - adds a single dimention to a target tensor

x_unsqueezed = torch.unsqueeze(x, dim=1)
x_unsqueezed
```

```python
# torch.permute - rearranges the dimentions of a target tensor in specified order 

x_original = torch.rand(size=(224, 224, 3)) # [height, width, color-channel]

# Permute 
x_permuted = x_original.permute(2, 0, 1) # shifts axis  0->1, 1->2, 2->0

x_original.shape, x_permuted.shape
```
